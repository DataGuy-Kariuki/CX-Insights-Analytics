```python
import pandas as pd
import time
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError

geolocator = Nominatim(user_agent="my_unique_user_agent")  # Use a unique user agent
cache = {}  # Dictionary to store lat/lon -> county mappings

def get_county(lat, lon):
    # Check if we already have the county for these coordinates
    key = (lat, lon)
    if key in cache:
        return cache[key]  # Return from cache

    try:
        loc = geolocator.reverse(f"{lat},{lon}", addressdetails=True)

        if loc is None or 'address' not in loc.raw:
            county = "Unknown County"
        else:
            address = loc.raw['address']
            county = address.get("county") or address.get("state") or "Unknown County"

        cache[key] = county  # Save to cache
        return county

    except (GeocoderTimedOut, GeocoderServiceError) as e:
        print(f"Geocoding error: {e}")
        return "Error"
```

```python
# get the loadtions
# Load user locations (assuming DataFrame `user_location` exists)
unique_coords = user_location[['latitude', 'longitude']].drop_duplicates()

# Get counties for unique lat/lon pairs
unique_coords['county'] = unique_coords.apply(lambda row: get_county(row['latitude'], row['longitude']), axis=1)

# Merge county back to the main dataset
user_location = user_location.merge(unique_coords, on=['latitude', 'longitude'], how='left')

# Save to CSV
user_location.to_csv("customer_locations.csv", index=False)

print("County mapping complete!")
```
